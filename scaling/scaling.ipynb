```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("Social_Network_Ads.csv")
df.head()
```

```python
df.drop(['User ID','Gender'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
x=df.drop('Purchased',axis=1)
y=df['Purchased']
```

```python
# train data 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

# StandardScaler
- mean will be zero and standar=1

```python
scaled=StandardScaler()
x_train_scaled=scaled.fit_transform(x_train)
x_test_scaled=scaled.transform(x_test)
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
y_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(x_train['Age'], x_train['EstimatedSalary'])
ax1.set_title("Before Scaling")
ax2.scatter(x_train_scaled['Age'], x_train_scaled['EstimatedSalary'],color='red')
ax2.set_title("After Scaling")
plt.show()
```

# min-max scalar
- Range 0 to 1

```python
min_max_scaled=MinMaxScaler()
x_train_scaled=min_max_scaled.fit_transform(x_train)
x_test_scaled=min_max_scaled.transform(x_test)
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(x_train['Age'], x_train['EstimatedSalary'])
ax1.set_title("Before Scaling")
ax2.scatter(x_train_scaled['Age'], x_train_scaled['EstimatedSalary'],color='green')
ax2.set_title("After Scaling")
plt.show()
```
