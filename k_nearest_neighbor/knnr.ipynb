```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
df=pd.read_csv('zameen-updated.csv')
df.head()
```

```python
df.shape
```

```python
# sample data if you implement all data size (168446, 20) then you can implement but I only 1000 rows implement 
# because big data and not support my pc 
df=df.head(1000)
```

```python
df.isnull().sum()/len(df)*100
```

```python
df.dtypes
```

```python
df['agency']=df['agency'].fillna(df['agency'].mode()[0])
df['agent']=df['agent'].fillna(df['agent'].mode()[0])
```

```python
df.drop('page_url',axis=1,inplace=True)
```

```python
for col in df.columns:
    df[col]=LabelEncoder().fit_transform(df[col])
df.head()
```

```python
x=df.drop('price',axis=1)
y=df['price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
knn=KNeighborsRegressor()
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
print('r2_score :',r2_score(y_test,y_pred))
print('mean_absolute_error :',mean_absolute_error(y_test,y_pred))
print('mean_squared_error :',mean_squared_error(y_test,y_pred))
```
